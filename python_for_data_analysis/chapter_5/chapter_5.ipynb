### Reindexing

### Arithmetic and Data Alignment

```python
s1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=['a', 'c', 'd', 'e'])
s2 = pd.Series([-2.1, 3.6, -1.5, 4, 3.1],
               index=['a', 'c', 'e', 'f', 'g'])
s1
```

```python
s2
```

```python
s1 + s2
```

```python
# index alignment performed on both row and columns
df1 = pd.DataFrame(np.arange(9.).reshape((3, 3)),
                   columns=list('bcd'),
                   index=['Ohio', 'Texas', 'Colorado'])
df2 = pd.DataFrame(np.arange(12.).reshape((4, 3)),
                   columns=list('bde'),
                   index=['Utah', 'Ohio', 'Texas', 'Oregon'])
df1
```

```python
df2
```

```python
# resulting index & columns are the unions of the ones in each DataFrame
df1 + df2
```

```python
df1 = pd.DataFrame({'A': [1, 2]})
df2 = pd.DataFrame({'B': [3, 4]})
df1
```

```python
df2
```

```python
df1 - df2 # produces all NaNs, since no columns or row labels are in common
```

```python
df1 = pd.DataFrame(np.arange(12.).reshape((3, 4)),
                   columns=list('abcd'))
df2 = pd.DataFrame(np.arange(20.).reshape((4, 5)),
                  columns=list('abcde'))
df2.loc[1, 'b'] = np.nan
df1
```

```python
df2
```

```python
df1 + df2
```

```python
df1.add(df2, fill_value=0)
```

```python
1 / df1
```

```python
df1.rdiv(1) # equivalent
```

```python
df1.reindex(columns=df2.columns, fill_value=0)
```

### Functions Application and Mapping

```python
frame = pd.DataFrame(np.random.randn(4, 3), columns=list('bde'),
                     index=['Utah', 'Ohio', 'Texas', 'Oregon'])
frame
```

```python
np.abs(frame)
```

```python
f = lambda x: x.max() - x.min()
frame.apply(f)
```

```python
frame.apply(f, axis='columns')
```

```python
# Apply accepts functions that return a Series with multiple values
def f(x):
    return pd.Series([x.min(), x.max()], index=['min', 'max'])

frame.apply(f)
```

```python
# Element-wise Python functions can be used, too
format = lambda x: '%.2f' % x
frame.applymap(format)
```

```python
frame['e'].map(format)
```

### Sorting and Ranking

```python
obj = pd.Series(range(4), index=['d', 'a', 'b', 'c'])
obj.sort_index()
```

```python
frame = pd.DataFrame(np.arange(8).reshape((2, 4)),
                     index=['three', 'one'],
                     columns=['d', 'a', 'b', 'c'])
frame.sort_index()
```

```python
frame.sort_index(axis=1)
```

```python
frame.sort_index(axis='columns') # equivalent
```

```python
frame.sort_index(axis=1, ascending=False)
```

```python
obj = pd.Series([4, 7, -3, 2])
obj.sort_values()
```

```python
# Missing values are sorted to the end by default
obj = pd.Series([4, np.nan, 7, np.nan, -3, 2])
obj.sort_values()
```

```python
frame = pd.DataFrame({'b': [4, 7, -3, 2], 'a': [0, 1, 0, 1]})
frame
```

```python
frame.sort_values(by='b')
```

```python
frame.sort_values(by=['a', 'b']) # 'b' is secondary sort here
```

```python
obj = pd.Series([7, -5, 7, 4, 2, 0, 4])
obj.rank()
```

```python
obj.rank(method='first')
```

```python
obj.rank(ascending=False, method='max')
```

```python
frame = pd.DataFrame({'b': [4.3, 7, -3, 2],
                      'a': [0, 1, 0, 1],
                      'c': [-2, 5, 8, -2.5]})
frame
```

```python
frame.rank(axis='columns')
```
